```r
download_daily_climatology <- function(start_year = 1998, end_year = 2023) {
  all_data <- list()
  
  for (year in start_year:end_year) {
    cat("Downloading data for year:", year, "\n")
    
    # Retrieve data for the year
    yearly_data <- aemet_daily_clim(
      start = paste0(year, "-01-01"), 
      end = paste0(year, "-12-31")
    )
    
    all_data[[as.character(year)]] <- yearly_data
    
    # Pause between requests to respect API limits (optional)
    Sys.sleep(1) 
  }
  
  # Combine all yearly data into a single data frame
  combined_data <- do.call(rbind, all_data)
  
  return(combined_data)
}
```

```r
# Download the data from 1998 to 2023
climatology_data <- download_daily_climatology(1998, 2023)
```
